```python
import numpy as np
import copy
```

```python
def Richardison(A,c,k):
    I = np.identity(len(A))
    B = (I-A)
    #print('A \n',A,'\n')
    #print('I \n',I,'\n')
    #print('B \n',B,'\n')
    #print('c \n',c,'\n')
    NB = np.max(np.abs(B))
    print('NB \n',NB,'\n')
    x = np.zeros(len(c))
    if NB >= 1: print('N√£o garante a converg√™ncia','NB=',NB,'B=',B)
    for i in range(k):
        xOld = copy.deepcopy(x) 
        x = np.dot(B,x) + c
        Er = np.max(np.abs(xOld-x))
        print('x= ',x,'Er=',Er,'it=',i)
        
```

```python
A = np.array([[1.2,0.5],[0.3,0.8]])
c = np.array([-2.6,0.7])
Richardison(A,c,20)

A = np.array([[1.5,1],[3,1.4]])
c = np.array([-2.5,2.2])
Richardison(A,c,100)

A = np.array([[1.5,1],[0.2,1.4]])
c = np.array([-2.5,-6.2])
Richardison(A,c,100)
```
