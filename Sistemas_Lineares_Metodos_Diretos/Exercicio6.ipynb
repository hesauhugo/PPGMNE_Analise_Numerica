```python
import numpy as np

def Gauss(A,b):
    n,c = A.shape
    for j in range(n):
        pivo = A[j][j]
        for i in range(j+1,n):
            fator = np.round(A[i][j]/pivo,4)
            print("Fator Gaus",fator)
            A[i][:]= np.round(A[i][:]  - fator*A[j][:],4)
            b[i]= np.round(b[i]- fator*b[j],4)
    return A,b

def triangularSuperior(k,n,x,a,b):
    x[n-1] = np.round(b[n-1]/a[n-1][n-1],4)
    print(x[n-1],n-1)
    k = n-2
    while k >= 0:
        s=0
        for j in range(k+1,n):
            s= np.round(s + a[k][j]*x[j],4)
            print("k=",k,"j=",j,"a[k][j]",a[k][j],"x[j]",x[j],"n-1 = ",n-1)
        x[k] = np.round((b[k]-s)/a[k][k],4)
        print("k=",k,"xk = ",x[k],"bk= ",b[k],"s= ",s,"akk= ",a[k][k])
        k= k-1
    return x
```

#### Resposta a)

```python
A = np.array([
    [10e-4,10e-4],
    [20,1]
    ])
b = np.array([10e-4,21])
A ,b = Gauss(A,b)
```

```python
print(A ,"\n",b)
```

```python
for a in A:
    print("a= ",a)

for _b in b:
    print("b= ",_b)
    
k,n = A.shape
x = np.zeros(n)
x = triangularSuperior(k,n,x,A,b)  
print("resultado a) = ",x)
```

#### Resultado a) =  [ 1.1    -0.0526]
