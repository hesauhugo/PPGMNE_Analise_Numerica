```python
def f(x):
    return (-7/12)*x**2 + (33/4)*x + (13/6)
```

```python
def h_trapezio(a,b,m):
    x = []
    y= []
    _h = (b-a)/m
    x.append(a)
    y.append(f(a))
    for i in range(1,m):
        a += _h
        x.append(a)
        y.append(f(a))
    x.append(b)
    y.append(f(b))
    return x,y
```

```python
a=1
b=5
m=8
x,y = h_trapezio(a,b,m)
```

```python
print(x)
print(y)
```

```python
def trapezioRepitida(x,a,b,m):
    h = (b-a)/m
    s = 0
    for i in range(1,m):
        s += f(x[i])
        print("i = ",i,"fx=",f(x[i]))
    v = (h/2)*(f(x[0]) + 2*s + f(x[m]))
    print("i=0",f(x[0]))
    print("i=m",f(x[m]))
    print("m = ",(m),"m =",m ,"x=",x,"h=",h)
    return v
```

```python
tresult = trapezioRepitida(x,a,b,m)
tresult
```

```python
def h_simpson(a,b,m):
    
    x = []
    y= []
    c= []
    _h = (b-a)/m
    x.append(a)
    y.append(f(a))
    c.append(1)
    
    for i in range(1,m):
        a += _h
        x.append(a)
        y.append(f(a))
        
        if i % 2 == 0:
            c.append(2)
        else:
            c.append(4)
        
    x.append(b)
    y.append(f(b))
    c.append(1)
    return x,y,c
        
```

```python
a=3
b=5
m=4
x,y,c = h_simpson(a,b,m)
print(x)
print(y)
print(c)
```

```python
def simpsonRepetida(x,c,y,a,b,m):
    
    h = (b-a)/m
    s = 0
    
    for i in range(len(x)):
        s += c[i]*y[i]
        print('x[i]',x[i],'c[i] = ',c[i],'y[i] = ', y[i])

    v = (h/3)*s
    return v
```

```python
Sresult = simpsonRepetida(x,c,y,a,b,m)
Sresult
```
