```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
def f(x):
    return (x**2)*math.exp(-x**2)
```

```python
def h(a,b,m):
    
    x = []
    y= []
    c= []
    _h = (b-a)/m
    x.append(a)
    y.append(f(a))
    c.append(1)
    
    for i in range(1,m):
        a += _h
        x.append(a)
        y.append(f(a))
        
        if i % 2 == 0:
            c.append(2)
        else:
            c.append(4)
        
    x.append(b)
    y.append(f(b))
    c.append(1)
    return x,y,c
        
```

```python
a=0
b=2
m=6
x,y,c = h(a,b,m)
print(x)
print(y)
print(c)
```

```python
xx = np.linspace(a,b)
yy =  (xx**2)*np.exp(-xx**2)
plt.plot(x,y,'-',label='Simpson')
plt.plot(xx,yy,'r',label='f(x)')
plt.legend()
```

```python
def simpsonRepetida(x,c,y,a,b,m):
    
    h = (b-a)/m
    s = 0
    
    for i in range(len(x)):
        s += c[i]*y[i]
        print('x[i]',x[i],'c[i] = ',c[i],'y[i] = ', y[i])

    v = (h/3)*s
    return v
```

```python
Sresult = simpsonRepetida(x,c,y,a,b,m)
Sresult
```
