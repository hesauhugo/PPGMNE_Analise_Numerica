```python
import numpy as np
```

```python
a= np.array([[4,3,0],[3,4,-1],[0,-1,4]])
k,n = a.shape
for i in range(n-1):
    determinante = np.linalg.det(a)
    if determinante > 0:
        print(determinante,">0")
    else:
        print(determinante,"<=0")
    a = a[:-1,:-1]
    print(a)
```
