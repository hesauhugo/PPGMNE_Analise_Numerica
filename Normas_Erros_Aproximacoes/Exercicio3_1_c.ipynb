##### Polinômio de Taylor
1. Construa o polinômio de Taylor n de grau  para a funções abaixo, em torno do ponto x0 indicado

c) $f(x)  = \sqrt{x} \text{ , } {x}_{0} \text{= 1}$ 

```python
from sympy import *
from sympy.plotting import *


%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

a,x,y = symbols('a x y')
y= sqrt(x)
```

```python
plot(y)
```

```python
primeiraDerivada = diff(sqrt(x))
resultadoPrimeiraDerivada = primeiraDerivada.subs(x,1)
print(primeiraDerivada)
print(resultadoPrimeiraDerivada)
primeiraDerivada
```

```python
plot(y,1/(2*sqrt(x)))
```

```python
segundaDerivada = diff(1/(2*sqrt(x)))
resultadoSegundaDerivada = segundaDerivada.subs(x,1)
print(segundaDerivada)
print(resultadoSegundaDerivada)
segundaDerivada
```

```python
plot(y,1/(2*sqrt(x)),1/(4*x**(3/2)))
```

```python
terceiraDerivada = diff(-1/(4*x**(3/2)))
resultadoTerceiraDerivada = terceiraDerivada.subs(x,1)
print(terceiraDerivada)
print(resultadoTerceiraDerivada)
terceiraDerivada
```

```python
quartaDerivada = diff(0.375*x**(-2.5))
resultadoQuartaDerivada = quartaDerivada.subs(x,1)
print(quartaDerivada)
print(resultadoQuartaDerivada)
quartaDerivada
```
