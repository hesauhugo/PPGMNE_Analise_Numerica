```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
def f(x):
    return ((x)*math.exp(2*x))/(1+2*x)**2
```

```python
def h(a,b,m):
    x = []
    y= []
    _h = (b-a)/m
    x.append(a)
    y.append(f(a))
    for i in range(1,m):
        a += _h
        x.append(a)
        y.append(f(a))
    x.append(b)
    y.append(f(b))
    return x,y
```

```python
a=0
b=3
m=6
x,y = h(a,b,m)
```

```python
xx = np.linspace(a,b)
yy = ((xx)*np.exp(2*xx))/(1+2*xx)**2
plt.plot(x,y,'-',label='Tr√°pezio')
plt.plot(xx,yy,'r',label='f(x)')
plt.legend()
```

```python
def trapezioRepitida(x,a,b,m):
    h = (b-a)/m
    s = 0
    for i in range(1,m):
        s += f(x[i])
        print("i = ",i,"fx=",f(x[i]))
    v = (h/2)*(f(x[0]) + 2*s + f(x[m]))
    print("i=0",f(x[0]))
    print("i=m",f(x[m]))
    print("m = ",(m),"m =",m ,"x=",x,"h=",h)
    return v
```

```python
tresult = trapezioRepitida(x,a,b,m)
tresult
```
