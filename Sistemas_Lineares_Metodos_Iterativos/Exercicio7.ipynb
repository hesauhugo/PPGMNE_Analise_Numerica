### Jacobi

$
\text{O método de Jacobi pode ser escrito em uma matriz com a seguinte notação:} \\
\mathit{Dx^{(k+1)} + (L+U)x^{(k)} = b} \\
\mathit{x^{(k+1)} = D^{-1}[(-L-U)x^{k} + b]}
$

$
\mathit{x_i^{(k+1)} = x_i^k} + \frac{1}{a_{ii}} \left( b_i - \sum\limits_{j=1}^{i-1} a_{ij} x_j^{(k)}
 - \sum\limits_{j=i}^{n} a_{ij} x_j^{(k)}
\right)
$

### Gauss-Seidel

 $
\text{O método de Gauss-Seidel pode ser escrito em uma matriz com a seguinte notação:} \\
\mathit{(L+D)x^{k+1} + Ux^{(k)} = b} \\
\mathit{x^{(k+1)} = (L+D)^{-1} [-Ux^{(k)} +b] }
$

$
\mathit{x_i^{(k+1)} = x_i^k} + \frac{1}{a_{ii}} \left( b_i - \sum\limits_{j=1}^{i-1} a_{ij} x_j^{(k+1)} - \sum\limits_{j=i}^{n} a_{ij} x_j^{(k)} \right)
$

### SOR

$
\mathit{Dx^{(k+1)} = b -Lx^{(k+1)} -Ux^{(k)}}
$

$
\mathit{x^{(k+1)} = D^{-1}[b - Lx^{(k+1)} - Ux^{(k)}]}
$

$
\mathit{x^{(k+1)} - x^{(k)} = D^{-1}[b - Lx^{(k+1)} - Dx^{(k)} - Ux^{(k)}]}
$

$
\mathit{x^{(k+1)} = x^{(k)} +\omega(x^{(k+1)} - x^{(k)})_{GS}}
$

$
\mathit{(x^{(k+1)} - x^{(k)})_{GS} =  D^{-1}[b - Lx^{(k+1)} - Dx^{(k)} - Ux^{(k)}]}
$

$
\mathit{x^{(k+1)} = x^{(k)} + \omega D^{-1}[b - Lx^{(k+1)} - Dx^{(k)} - Ux^{(k)}]} 
$

$
\mathit{ (L + \frac{1}{\omega} D)x^{(k+1)} = \frac{1}{\omega} Dx^{(k)} + [b - Dx^{(k)} - Ux^{(k)}] = (\frac{1}{\omega} -D -D - U)x^{(k)} + b} 
$

$
\mathit{ x^{(k+1)} = (L + \frac{1}{\omega} D)^{-1} [(\frac{1}{\omega} -D -D - U)x^{(k)} + b]} 
$
