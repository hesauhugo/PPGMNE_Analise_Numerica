```python
import numpy as np
import copy
```

```python
def MatrizB(A):
    AIdent = np.identity(len(A))
    D = copy.deepcopy(A)*copy.deepcopy(AIdent)
    #print(D)
    InvD = np.linalg.inv(copy.deepcopy(D))
    #print(InvD)
    B = copy.deepcopy(AIdent) - np.dot(copy.deepcopy(InvD),copy.deepcopy(A))
    return B
```

```python
def Sassenfeld(B):
    m,n = B.shape
    Beta = np.zeros(m)

    for i in range(m):
        s = 0
        for j in range(n):
            if j < i:
                print("< i= ", i,"j= ",j,"Beta[j] = ",Beta[j],"B[i][j] = ", B[i][j])
                s += np.dot(np.abs(Beta[j]), np.abs(B[i][j]))
            elif j > i:
                print("> i= ", i,"j= ",j,"B[i][j]= ",B[i][j])
                s += np.abs(B[i][j])
        #print("Beta= ", Beta,"s= ",s)
        Beta[i] = s

    print("*Beta= ", Beta,"*s= ",s)
    BetaMax =np.max(Beta)   
    print("*BetaMax= ",BetaMax)
    if BetaMax<1:
        print("converge")
    else:
        print("n√£o converge")
```

```python
A = np.array([[4,1,-2],[2,3,0],[2,-2,6]])
B = MatrizB(A)
Sassenfeld(B)
```
