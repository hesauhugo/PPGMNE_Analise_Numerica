```python
 def f(x):
    return x**3 + 3*x**2 + 2*x -1
```

```python
def h(a,b,m):
    
    x = []
    y= []
    c= []
    _h = (b-a)/m
    x.append(a)
    y.append(f(a))
    c.append(1)
    
    for i in range(1,m):
        a += _h
        x.append(a)
        y.append(f(a))
        
        if i % 2 == 0:
            c.append(2)
        else:
            c.append(4)
        
    x.append(b)
    y.append(f(b))
    c.append(1)
    return x,y,c
```

```python
a=0
b=2
m=5
_x,_y,_c = h(a,b,m)
print(_x)
print(_y)
print(_c)
```

```python
xx = np.linspace(a,b)
yy = xx**3 + 3*xx**2 + 2*xx -1
plt.plot(_x,_y,'-',label='Simpson',marker = 'o')
plt.plot(xx,yy,'r',label='f(x)')
plt.legend()
```
