$
\text{(a) Estime quantas iterações do método da Bisseção }
\\ \text{serão necessárias para determinar } 
\\ \sqrt{3} \text{ com precisão }  \mathit{10}^{-4} 
\text{ no intervalo } \mathit{[1;2]}.
$

```python
import math
```

```python
def fx(x):
    return x**2 -3
```

```python
a=1
b=2
precisao = 10**-4
k = (math.log(b-a,10) - math.log(precisao,10)) / math.log(2,10)
k
```

```python
x = a
k = 20
it = 0

while True:
    it +=1
    #print("x= ", x,"iteracoes =", it,"f(x) = ",fx(x)) 
    xOld = x
    x=a+((b-a)/2)
    
    if fx(a)*fx(x) < 0:
        #print("x=b= ", x) 
        b=x
    else:
        #print("x=a= ", x) 
        a=x 
    if k == it:
        break
       
        
print("x= ", x,"iteracoes =", it,"f(x) = ",fx(x))
```

##### Resposta a:
$ k \approx \mathit{14} $ 
$
\\ \text{O resultado encontrado foi de } \mathit{x\approx 1.73199} \text{ para } \mathit{y \approx -0.00019}
