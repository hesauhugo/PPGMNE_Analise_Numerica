##### Polinômio de Taylor
3.Considere a função  definida por 

$ f(x) \text{ = } exp\left(\frac{4x}{5} \right)
\text{ + } exp\left(5-x \right)  $

(d) Repita o item anteior, mas agora considerando $x_0 = 2,5$.

```python
import math
```

```python
x=0
x0=2.5
N=100
s_n  = 0
resultado = []
info = []
for i in range(0,500):
    if i==0:
        x = 0
    else:
        x = i/100
    s_n  = 0    
    for n in range(1,N):
        s_n = s_n + (math.pow(4/5,n)*math.exp(4*x/5) + math.pow(-1,n)*math.exp(5-x))*math.pow((x-x0),n)
        erro = abs(math.exp(4*(x/5)) + math.exp(5-x) - s_n)
        if erro < 0.5:
            resultado.append(erro)
            info.append([x,n])
            print("x= ",x,"n = ",n, "erro = ",erro)
  
```

```python
indiceMenor = resultado.index(min(resultado))
indiceMenor
```
