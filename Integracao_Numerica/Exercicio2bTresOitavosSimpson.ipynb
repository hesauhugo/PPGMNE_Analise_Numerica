```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
def f(x):
    return ((x)*math.exp(2*x))/(1+2*x)**2
```

```python
def h(a,b,m):
    
    x = []
    y= []
    c= []
    _h = (b-a)/m
    x.append(a)
    y.append(f(a))
    c.append(1)
    contagem = 0
    
    for i in range(1,m):
        a += _h
        x.append(a)
        y.append(f(a))
        
        if contagem == 2:
            c.append(2)
            contagem = 0
        else:
            c.append(3)
            contagem += 1
        
    x.append(b)
    y.append(f(b))
    c.append(1)
    return x,y,c
        
```

```python
a=0
b=3
m=8
x,y,c = h(a,b,m)
print(x)
print(y)
print(c)
```

```python
xx = np.linspace(a,b)
yy = ((xx)*np.exp(2*xx))/(1+2*xx)**2
plt.plot(x,y,'-',label='Simpson',marker = 'o')
plt.plot(xx,yy,'r',label='f(x)')
plt.legend()
```

```python
def simpsonRepetida(x,c,y,a,b,m):
    
    h = (b-a)/m
    s = 0
    
    for i in range(len(x)):
        s += c[i]*y[i]
        print('x[i]',x[i],'c[i] = ',c[i],'y[i] = ', y[i])

    v = (3*h/8)*s
    return v
```

```python
Sresult = simpsonRepetida(x,c,y,a,b,m)
Sresult
```
