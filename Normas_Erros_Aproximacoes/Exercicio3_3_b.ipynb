##### Polinômio de Taylor
3.Considere a função  definida por 

$ f(x) \text{ = } exp\left(\frac{4x}{5} \right)
\text{ + } exp\left(5-x \right)  $

(b) Construa um programa que avalie $ T_{n}(x;x_{0})$

```python
import math
```

```python
def Taylor(x,x0,N):
    s_0 = math.exp((4*x)/5) + math.exp(5-x)
    s_n  = s_0
    for n in range(1,N):
        s_n = s_n + (math.pow(4/5,n)*math.exp(4*x/5) + math.pow(-1,n)*math.exp(5-x))*math.pow((x-x0),n)
        print(s_n)
```

```python
Taylor(2,1,100)
```
