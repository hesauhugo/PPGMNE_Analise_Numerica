##### Polinômio de Taylor
1. Construa o polinômio de Taylor de grau  para a funções abaixo, em torno do ponto 
 indicado

a) $f(x)  = {(1-x)}^{-1}, {x}_{0} = 0$ 

```python
from sympy import *
from sympy.plotting import *


%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

a,x,y = symbols('a x y')
y= (1-x)**-1
```

```python
xx = np.linspace(-10,10,10)
plt.figure(1)
y= (1-xx)**-1
plt.plot(xx,y,'-r',label = 'y')
plt.legend()
```

```python
primeiraDerivada = diff((1-x)**-1)
resultadoPrimeiraDerivada = primeiraDerivada.subs(x,0)
print(resultadoPrimeiraDerivada)
print(primeiraDerivada)
primeiraDerivada
```

```python
xx = np.linspace(-10,10,10)
plt.figure(1)
y= (1-xx)**-1
yd= (1 - xx)**(-2)
plt.plot(xx,y,'-r',label = 'y')
plt.plot(xx,yd,'-g',label = 'yd')
plt.legend()
```

```python
segundaDerivada = diff((1 - x)**-2)
resultadoSegundaDerivada = segundaDerivada.subs(x,0)
print(resultadoSegundaDerivada)
print(segundaDerivada)
segundaDerivada
```

```python
xx = np.linspace(-10,10,10)
plt.figure(2)
y= (1-xx)**-1
yd= (1 - xx)**(-2)
ydd = (2/(1 - xx)**3)
plt.plot(xx,y,'-r',label = 'y')
plt.plot(xx,yd,'-g',label = 'yd')
plt.plot(xx,ydd,'-',label = 'ydd')
plt.legend()
```

```python
terceiraDerivada = diff(2/(1 - x)**3)
resultadoTerceiraDerivada = terceiraDerivada.subs(x,0)
print(resultadoTerceiraDerivada)
print(terceiraDerivada)
terceiraDerivada
```

```python
xx = np.linspace(-10,10,10)
plt.figure(2)
y= (1-xx)**-1
yd= (1 - xx)**(-2)
ydd = (2/(1 - xx)**3)
yddd = (6/(1 - xx)**4)
plt.plot(xx,y,'-r',label = 'y')
plt.plot(xx,yd,'-g',label = 'yd')
plt.plot(xx,ydd,'-',label = 'ydd')
plt.plot(xx,ydd,'-y',label = 'yddd')
plt.legend()
```

```python
quartaDerivada = diff(6/(1 - x)**4)
resultadoQuartaDerivada = quartaDerivada.subs(x,0)
print(resultadoQuartaDerivada)
print(quartaDerivada)
quartaDerivada
```
